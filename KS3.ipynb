```python
import os
import sys
import pandas as pd
import numpy as np
from scipy.stats import ks_2samp
import matplotlib.pyplot as plt
```

```python
def get_ibs(firefly_array, steps=200000, stepsize=0.009):
    ibs = []
    current_time = 0.0
    last_time = 0.0
    stepsize = stepsize
    flash_array = np.array([False for i in range(steps)])
    for firefly in firefly_array:
        flash_array += np.array(firefly.flashed_at_this_step)
    for step in flash_array:
        current_time += stepsize
        if step:
            diff = current_time - last_time
            if diff > 2.0:
                ibs.append(diff)
            last_time = current_time
    return ibs
```

```python
def merge_ibs(data, steps=200000, stepsize=0.009):
    ibs = []
    for sim in data:
        ibs.extend(get_ibs(sim.firefly_array, steps, stepsize))
    return ibs
```

```python
files = os.listdir('C:/Users/peter/Documents/CU/23Fall/APPM 5370/Project/Clean/EmergentPeriodicity/Sandbox/05ff')
compare = pd.read_csv('../EmergentPeriodicity/experimental_data/experimental/ibs05ff.csv')
compare = compare['0'].values

ks_values = {}

for ind, file in enumerate(files):
    path = 'Sandbox/05ff/' + file
    print(path)
    data = pd.read_pickle(path)
    ibs = merge_ibs(data)
    if ind % 5 == 0:
        plt.hist(ibs, bins=100)
        plt.show()
    try:
        ks_value = ks_2samp(ibs, compare).statistic
        ks_values[round((ind + 1) / 100, 2)] = ks_value
    except:
        ks_values[round((ind + 1) / 100, 2)] = None
```

```python
plt.plot(list(ks_values.keys()), list(ks_values.values()))
```
